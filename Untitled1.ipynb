```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('NSCC_R2_Q4.csv')
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(15)
ax.plot(df['Wavelength'], df['Amount of light blocked'], lw=1, c='k', marker='o')
plt.xlabel('Wavelength in Micrometer')
plt.ylabel('Amount of light blocked in ppm')

#water 
ax.axvspan(2.5, 2.7, alpha=0.3, color='blue') 
ax.axvspan(1.86, 1.96, alpha=0.3, color='blue') 
ax.axvspan(1.38, 1.38, alpha=0.5, lw=3, color='blue') 
#co2 
# ax.axvspan(4.2, 4.4, alpha=0.3, color='red') 
ax.axvspan(2.6, 2.8, alpha=0.3, color='red')
ax.axvspan(1.9, 2.1, alpha=0.3, color='red') 

fig.suptitle('Transmission Spectra')
plt.tight_layout()
plt.savefig('Q4_plot.png')
plt.show()
```
