```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_C = pd.read_csv('C.csv')
```

```python
df_C['force_att'] = df_C['k2']/(df_C['R']**4) + 4*df_C['T']/df_C['R']
df_C['force_rep'] = df_C['k1']/(df_C['R']**3)
```

```python
plt.style.use('seaborn')
plt.plot(np.log10(df_C['force_att']), np.log10(df_C['force_rep']), "o", label="Data points")
x = np.arange(-2, 4)
y = np.arange(-2, 4)
plt.plot(x, y, 'k-', label="Equilibrium Boundary")
plt.xlabel("log(Attractive Pressure)")
plt.ylabel("log(Repulsive Pressure)")
plt.legend()
```
